```python
import urllib
from bs4 import BeautifulSoup
import pandas as pd

from matplotlib import pyplot as plt
import numpy as np 


url = 'https://en.wikipedia.org/wiki/Oklahoma'

page = urllib.request.urlopen(url)
soup = BeautifulSoup(page, "lxml")
soup.prettify()
soup.title

soup.title.string
```

```python
for row in soup.findAll('tr'):
    cells=row.findAll('td')
    print(len(cells))
for cell in cells:
    print(cell)
```

```python
A=[]
B=[]
C=[]

for row in soup.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==4:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
        
```

```python
df=pd.DataFrame(A,columns=['Census'])
df['Population']=B
df['Percent']=C


df_info = df.head(13)

B = df['Population'].head(13)

b = [x.replace(',', '') for x in B]
b_int = [int(y) for y in b]
```

```python
array = df_info.to_numpy
array
```

```python
census = df_info.Census
census_int = census.astype(int)
pop = b_int
```

```python
plt.figure(figsize=(10,10))
plt.title("Historical Population of Oklahoma")
plt.xlabel("Year")
plt.ylabel("Population per million")
plt.bar(range(0,13), pop, color='orange')
plt.xticks(range(0,13), census_int)

plt.show()
```
